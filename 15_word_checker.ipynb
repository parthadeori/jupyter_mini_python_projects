## **Advanced Word Counter that:**

1. Ignores punctuation
2. Is case-insensitive
3. Counts frequency of each word
4. Shows most repeated word

## **Step 1 — Logic**

**If user enters:**

```
Python is great. Python is easy!
```

**We want to treat:**

- `Python` and `python` as same
- Remove `.` and `!`
- Count how many times each word appears

Final cleaned words:

```
["python", "is", "great", "python", "is", "easy"]
```

---

## **Step 2 — Program**



````python
# Advanced Word Counter

import string

sentence = input("Enter a sentence: ")

# Convert to lowercase
sentence = sentence.lower()

# Remove punctuation
for char in string.punctuation:
    sentence = sentence.replace(char, "")

# Split into words
words = sentence.split()

# Count total words
print("Total words:", len(words))

# Word frequency
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print("\nWord Frequency:")
for word, count in word_count.items():
    print(word, ":", count)

# Most repeated word
most_word = max(word_count, key=word_count.get)
print("\nMost repeated word:", most_word)
````

## **Concepts Used**

- `string.punctuation`
- `for` loop
- dictionary
- `in` keyword
- `max()` with key
- `.lower()`

## **Final Program**

````python
import string

sentence = input("Enter a sentence: ")

sentence = sentence.lower()

# remove punctuation
for char in string.punctuation:
    sentence = sentence.replace(char, "")

words = sentence.split()

print("Total words:", len(words))

# word frequency
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print("\nWord Frequency:")
for word, count in word_count.items():
    print(word, ":", count)

# unique words
unique_words = set(words)
print("\nTotal unique words:", len(unique_words))

# top 3 words
sorted_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)

print("\nTop 3 most frequent words:")
for word, count in sorted_words[:3]:
    print(word, ":", count)
````
