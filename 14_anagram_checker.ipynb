## **What is an Anagram?**
Two words are anagrams if:
- They contain the same letters
- In the same frequency
- But possibly in different order

**Example:**

- listen → silent ✅
- triangle → integral ✅
- apple → papel ✅
- rat → car ❌

### **Simple Logic (Beginner Friendly)**
**Basic Idea:**
- Take two words as input
- Convert both to lowercase
- Sort both words
- Compare

*If sorted words are equal → Anagram*

*Else → Not anagram*

### **Version 1 – Using `sorted()` (Easy)**

```python
word1 = input("Enter first word: ")
word2 = input("Enter second word: ")

# Convert to lowercase
word1 = word1.lower()
word2 = word2.lower()

# Check if sorted words are equal
if sorted(word1) == sorted(word2):
    print("Anagram ✅")
else:
    print("Not an Anagram ❌")
```

### **Version 2 – Without `sorted()` (Better Logic)**

**Idea:**

Count frequency of each character manually.

```python
word1 = input("Enter first word: ").lower()
word2 = input("Enter second word: ").lower()

# If lengths are different, they cannot be anagrams
if len(word1) != len(word2):
    print("Not an Anagram ❌")
else:
    char_count = {}

    # Count letters in word1
    for char in word1:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    # Subtract letters using word2
    for char in word2:
        if char in char_count:
            char_count[char] -= 1
        else:
            print("Not an Anagram ❌")
            break

    # Final check
    for value in char_count.values():
        if value != 0:
            print("Not an Anagram ❌")
            break
    else:
        print("Anagram ✅")
```

### **Even Better Version (Clean & Professional)**

**Why This Is Better**

- Frequency counting → **O(n)**
- Better for very large strings

Interviewers LOVE this improvement.

```python
import unicodedata

def clean_text(text):
    # Convert to lowercase
    text = text.lower()

    # Normalize unicode (é → e)
    text = unicodedata.normalize('NFKD', text)
    text = ''.join(char for char in text if not unicodedata.combining(char))

    # Keep only letters and numbers
    text = ''.join(char for char in text if char.isalnum())

    return text


def is_anagram(text1, text2):
    text1 = clean_text(text1)
    text2 = clean_text(text2)

    # If lengths differ → impossible
    if len(text1) != len(text2):
        return False

    char_count = {}

    # Count characters from text1
    for char in text1:
        char_count[char] = char_count.get(char, 0) + 1

    # Subtract using text2
    for char in text2:
        if char not in char_count:
            return False
        char_count[char] -= 1
        if char_count[char] < 0:
            return False

    return True


# Testing
word1 = input("Enter first phrase: ")
word2 = input("Enter second phrase: ")

if is_anagram(word1, word2):
    print("Anagram ✅")
else:
    print("Not an Anagram ❌")
```
