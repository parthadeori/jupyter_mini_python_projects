## **What is a Palindrome?**
A palindrome is a word that reads the same forward and backward.

**Examples:**
- madam ✅
- racecar ✅
- hello ❌

### **Version 1 — Simple Palindrome Checker**

````python
word = input("Enter a word: ")

reversed_word = word[::-1]   # reverse using slicing

if word == reversed_word:
    print("It is a palindrome")
else:
    print("It is not a palindrome")
````

### **Version 2 — Using String Methods (More Practical)**

**Real-world inputs may have:**

- Capital letters
- Spaces
- Mixed cases

**Example:**

- Madam
- Race Car

````python
text = input("Enter text: ")

text = text.lower()          # make lowercase
text = text.replace(" ", "") # remove spaces

reversed_text = text[::-1]

if text == reversed_text:
    print("Palindrome")
else:
    print("Not Palindrome")
````

### **HARDER Version (Without Slicing)**

````python
text = input("Enter text: ")
text = text.lower().replace(" ", "")

reversed_text = ""

for char in text:
    reversed_text = char + reversed_text

if text == reversed_text:
    print("Palindrome")
else:
    print("Not Palindrome")
````

### **Clean, interview-ready function that:**
- Ignores uppercase/lowercase
- Ignores spaces
- Ignores punctuation
- Works for full sentences

````python
def is_palindrome(text):
    cleaned_text = ""

    # Keep only letters and numbers
    for char in text:
        if char.isalnum():          # keeps only a-z, A-Z, 0-9
            cleaned_text += char.lower()

    # Reverse
    reversed_text = cleaned_text[::-1]

    return cleaned_text == reversed_text


# Testing
user_input = input("Enter text: ")

if is_palindrome(user_input):
    print("Palindrome ✅")
else:
    print("Not Palindrome ❌")
````

### **Why This Is Better**

`isalnum()`

**Removes:**
- spaces
- commas
- periods
- special characters

`lower()`
- Makes comparison case-insensitive.

### **More Advanced (Two-Pointer Method)**

````python
def is_palindrome(text):
    left = 0
    right = len(text) - 1

    while left < right:
        if not text[left].isalnum():
            left += 1
            continue

        if not text[right].isalnum():
            right -= 1
            continue

        if text[left].lower() != text[right].lower():
            return False

        left += 1
        right -= 1

    return True
````

### **Why this is powerful:**
- No extra string created
- O(1) space complexity
- Cleaner logic
- Looks impressive in interviews

**1. Why is `isalnum()` better than `replace()` here?**

If we use `replace()`:
```
text = text.replace(" ", "")
text = text.replace(",", "")
text = text.replace(".", "")
```

**Problem:**
- You must manually remove EVERY special character.
- What about `! @ # $ % ^ & * ( ) ?`
- You’ll forget something.

---

**`isalnum()` does this automatically:**

```
char.isalnum()
```
**It keeps:**
- a–z
- A–Z
- 0–9

**It removes:**
- spaces
- punctuation
- symbols

Much cleaner. Much safer.

So in interviews → `isalnum()` = professional solution.
